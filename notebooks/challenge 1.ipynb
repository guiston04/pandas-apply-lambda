```python
def warm_up(x):
    if x >= 0 and x <=999:
        return 1
    elif x>=1000 and x<=9999:
        return 2
    elif x>=10000 and x<=99999:
        return 3
    elif x>=100000 and x<=999999:
        return 4
    else:
        return 5
```

```python
movies['bin'] = movies.apply(lambda x: warm_up( x['Votes']), axis=1 )
movies.head()
```

```python
movies['revenue per minute'] = movies['Revenue (Millions)'] / movies['Runtime (Minutes)']
```

```python
def ranking(x,y):
    if 'Comedy' in y:
        return int(x-1)
    elif 'Thriller' in y:
        return int(x+1)
    else:
        return x
```

```python
movies['new_ranking'] = movies.apply(lambda x: ranking(x['Rank'],x['Genre']), axis=1 )
```

```python
movies
```

```python
def real_stuff(x, y):
    n = int(sum([ord(i) for i in x]) / y)
    if n == 0: return False
    for i in range(2,n):
        if (n%i) == 0:
            return False
    return True
```

```python
movies['real stuff'] = movies.apply(lambda x: real_stuff(x['Title'], x['Votes']), axis=1)
movies.head()
```
